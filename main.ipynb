```python
word_list = word_list_str.split("\n")
word_idx = []
error_idx = []

OUT_FILE_PATH = "/Users/user/OSS/Test/part_of_speech.jsonl"
ERROR_FILE_PATH = "/Users/user/OSS/Test/error_word.jsonl"
OUT_FILE_CSV = "/Users/user/OSS/Test/element_student_words_800.csv"
```

```python
def save_part_of_words(word: str, outfile_path: str, error_files_path: str):
    part_of_words = []
    res = httpx.get(f"https://api.dictionaryapi.dev/api/v2/entries/en/{word}")
    if res.status_code == 200:
        res = res.json()
        for i in res:
            meanings = i["meanings"]
            for meaning in meanings:
                if meaning["partOfSpeech"] not in part_of_words:
                    part_of_words.append(meaning["partOfSpeech"])
        with open(outfile_path, 'a') as outfile:
            json.dump({"word": word, "part_of_speech": part_of_words}, outfile)
            outfile.write("\n")
    else :
        with open(error_files_path, 'a') as error_files:
            json.dump({"word": word, "error": res.status_code}, error_files)
            error_files.write("\n")

for word in word_list:
    print(word)
    save_part_of_words(word=word, outfile_path=OUT_FILE_PATH, error_files_path=ERROR_FILE_PATH)
    # t = threading.Thread(target=save_part_of_words, args=(word, outfile, error_file))
    # t.start()
    
```

```python

# id 정보를 모아서 저장하는 예시
col = []
idx = []
values = []

# open 내의 디렉토리 및 파일 이름에 유의
with jsonlines.open(OUT_FILE_PATH) as f:
    for line in f:
        for part_of_speech in line['part_of_speech']:
            if part_of_speech not in col:
                col.append(part_of_speech)

print(col)
```

```python
with jsonlines.open(OUT_FILE_PATH) as f:
    for line in f:
        row = [0 for i in range(len(col))]
        ARR_PART_OF_SPEECH = line['part_of_speech']
        WORD = line['word']
        for part_of_speech in ARR_PART_OF_SPEECH:
            ARR_INDEX = col.index(part_of_speech)
            row[ARR_INDEX] = 1
        idx.append(WORD)
        values.append(row)
```

```python
df = pd.DataFrame(values, columns=col, index=idx)
df.to_csv(path_or_buf=OUT_FILE_CSV)
```
